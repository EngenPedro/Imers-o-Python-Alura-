Importando a Biblioteca Python

```python
import pandas as pd
```

```python
df_principal = pd.read_excel('/content/Imersão Alura Tabela de Ações (1).xlsx', sheet_name='Principal')
df_principal.head(5)
```

```python
df_total_acoes = pd.read_excel('/content/Imersão Alura Tabela de Ações (1).xlsx', sheet_name='Total_de_acoes')
df_total_acoes
```

```python
df_chatgpt = pd.read_excel('/content/Imersão Alura Tabela de Ações (1).xlsx', sheet_name= 'Chat - GPT')
df_chatgpt = df_chatgpt.rename(columns={'Nome da Empresa': 'Nome da Empresa_Gpt','Segmento' : 'Segmento_Gpt'})
df_chatgpt
```

```python
df_ticker = pd.read_excel('/content/Imersão Alura Tabela de Ações (1).xlsx', sheet_name= 'Ticker')
df_ticker
```

```python
df_analises = pd.read_excel('/content/Imersão Alura Tabela de Ações (1).xlsx', sheet_name= 'Analises')
df_analises
```

```python
df_principal = df_principal.rename(columns={'Último (R$)' : 'Valor_Final', 'Var. Dia ' : 'Var_Dia_Pct'}).copy()
df_principal
```

```python
df_principal['Var_pct'] = df_principal['Var_Dia_Pct'] / 100
df_principal
```

```python
df_principal['Valor_inicial'] = df_principal['Valor_Final'] / (df_principal['Var_pct'] +1 )
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['Variacao_R$'] = (df_principal['Valor_Final'] - df_principal['Valor_inicial'])*df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
df_principal
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica' : 'Qtd_teorica'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['Variacao_R$'].apply(lambda x: "Subiu" if x > 0 else ("Desceu" if x < 0 else "Estável"))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome da Empresa_Gpt', how='left')
df_principal
```

```python
dx = df_principal.loc[:,~df_principal.columns.duplicated()]
dx
```

```python
df_principal['Cat_idade'] = df_principal['Idade em Anos'].apply(lambda x: "Mais de 100" if x > 100 else ("Menos de 50" if x < 50 else "Entre 50 e 100"))
df_principal
```

```python
maior_variacao = df_principal['Variacao_R$'].max()

# Menor Variação
menor_variacao = df_principal['Variacao_R$'].min()

# Média
media_variacao = df_principal['Variacao_R$'].mean()

media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_R$'].mean()

# Média de Quem Desceu
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_R$'].mean()
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

Agrupar por segmento e somar a variação

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_R$'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_R$'].sum().reset_index()
df_analise_saldo
```

```python
import plotly.express as px
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_R$', text='Variacao_R$', title='Variação em Reais por Resultado')
fig.update_traces(texttemplate='R$ %{text:,.2f}', textposition='outside')
fig.show()
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_R$'].sum().reset_index()
df_analise_segmento
```

# Utilizando o Copilot para melhorar o Gráfico de Pizza, adicionando a porcentagem das catagorias e colocando as em ordem crescente.

cop.png

```python
fig = px.pie(df_analise_segmento, names='Segmento', values='Variacao_R$', title='Variação Reais por Segmento')
fig.show()

# Adicionando rótulos de porcentagem
fig.update_traces(textinfo='percent+label')

# Ordenando os segmentos
fig.update_layout(
    updatemenus=[
        dict(
            buttons=list([
                dict(
                    args=[{"values": [sorted(fig.data[0]['values'])],
                           "labels": [sorted(fig.data[0]['labels'])]}],
                    label="Ordenar",
                    method="restyle"
                ),
                dict(
                    args=[{"values": [fig.data[0]['values']],
                           "labels": [fig.data[0]['labels']]}],
                    label="Desordenar",
                    method="restyle"
                )
            ]),
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1.1,
            yanchor="top"
        ),
    ]
)

fig.show()
```

```python
df_analise_cat_idade = df_principal.groupby('Cat_idade')['Variacao_R$'].sum().reset_index()
df_analise_cat_idade
```
